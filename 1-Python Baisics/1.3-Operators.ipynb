```python
##Arithmetic Operators
a=10
b=5
add_result=a+b ##Addition
subt_result=a-b  ##Subtraction
mult_result=a*b  ##Multiplication
div_result=a/b  ##Division
floor_div_result=a//b ##Floor Division
expo_result=a**b ##Exponentiation
mod_result=a%b ##Modulus

print(add_result)
print(subt_result)  
print(mult_result)
print(div_result)
print(floor_div_result)
print(expo_result)
print(mod_result)
```

```python
##Comparison Operators
a=10
b=10
print(a==b)  ##Equal to
print(a!=b)  ##Not equal to
```

```python
##Logical Operators
##AND, OR, NOT
X=True
Y=True
result=X and Y
print(result)
```

```python
##OR
X=False
Y=False
result=X or Y
print(result)
```

```python
##not Operator
X=True
not X
```
