```python
with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
##Read line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ##s strip removes new line character
```

```python
## Writing a file(Overwriting)
with open('example.txt','w') as file:
    file.write("Hello World\n")
    file.write("This is a new line\n")
```

```python
## Write a file without overwriting
with open('example.txt','a') as file:
    file.write("This is appended line")
    file.write("This is another appended line\n")
```

```python
## Writing a list of lines in a file
lines=["Line 1\n","Line 2\n","Line 3\n"]
with open('example.txt','w') as file:
    file.writelines(lines)
```

```python
## Binary Files
data=b'\x00\x01\x02\x03\x04\x05'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
## Reading Binary Files

with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
## Read the content from source text file nad write to a destination text file
#Copying a text file
with open('example.txt','r') as src_file:
    content=src_file.read()

with open('copy_example.txt','w') as dest_file:
    dest_file.write(content)
```

```python
## Read a text file and count the number of lines, words and characters
with open('example.txt','r') as file:
    Lines=file.readlines()
    line_count=len(Lines)
    words=0
    characters=0

    for line in Lines:
        print(line.split())
        words+=len(line.split())
        for char in line.split():
            characters+=len(char)
    print(line_count,words,characters)
   




```

```python
## Writing and then reading a file
with open('example.txt','w+') as file:
    file.write("Hello World\n")
    file.write("This is Dhruvi 778\nyoyo")
    ## Move the file cursor to the beginning
    file.seek(0)
    content=file.read()
    print(content)
```
