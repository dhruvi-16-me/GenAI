```python
## Inheritance
## Parent Class
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f"The person will drive the {self.enginetype} car" )
```

```python
car1=Car(4,5,"petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,autopilot):
        super().__init__(windows,doors,enginetype)
        self.autopilot = autopilot

    def autopilot_feature(self):
        print(f"Tesla supports autopilot : {self.autopilot}")
```

```python
tesla1=Tesla(4,5,"electric",True)
tesla1.autopilot_feature()
tesla1.drive()
```

```python
##Multiple Inheritance
##When a class inherits more than one base class
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Subclass must implement this method")

## Base class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner

## Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner,breed):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
        self.breed=breed

    def speak(self):
        return f"{self.name} says Woof!"
    
dog1=Dog("Buddy","Alice","Golden Retriever")
print(dog1.speak())
print(f"Owner: {dog1.owner}")
```
